```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import os 
os.listdir()
```

```python
df = pd.read_csv('Speed Dating Data.csv',encoding='ISO-8859-1')
```

```python
cols = np.sum(df.isnull()) < 0.10*len(df)
```

```python
df[cols[cols == True].keys().values].corr()
```

```python
for col in df.columns:
    print("column - ", col, " value - ", df[!df[col].isnull()][col])
```

```python
wave_freq = []
for sub_df in df.groupby(by='wave'):
    wave_freq.append((len(sub_df[1]), sub_df[0]))
```

```python
sorted(wave_freq)
```

```python
df_corr = df[df.wave==18].corr()
```

```python
for col in df_corr:
    print("column - ", col)
    print(df_corr[(df_corr[col] > 0.6) & (df_corr[col] < 1)][col])
    print("=========================")
```

```python
a = ["attr", "sinc", "intel", "fun", "amb"]
wave = [1, 8, 17, 12, 19, 7, 2, 4, 15, 14, 9, 11, 21]
for j in wave:
    print("WAVE ------------------------------ ", j)
    for i in range(1, 6):
        print([col+str(i)+"_1" for col in a])
        temp_df = df[[col+str(i)+"_1" for col in a] + ["wave"]]
        print("for qiestion - ", i, "number of nulls - ", np.sum(temp_df[temp_df.wave == j].isnull()))
    print("========================================")
```
